En total son 695 datos, existen columnas más del 50% de los datos faltantes. Para evitar alterar la integridad de los datos de esas columnas a la hora de aplicar promedios y/o modas, se descartarán aquellas columnas que tengan más del 50% de los datos faltantes.

```python
for column in data.columns:
  faltante = data[column].isnull().sum() * 100 / data['Carrera profesional cursada'].count()
  if faltante > 50:
    data = data.drop([column], axis=1)
```
